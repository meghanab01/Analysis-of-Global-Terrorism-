### Let's start cleaning up the data types

```python
gtd['date'] = pd.to_datetime(
    gtd[['year', 'month', 'day']].rename(columns={'year': 'year', 'month': 'month', 'day': 'day'}),
    errors='coerce'
)
```

```python
cat_cols = [
    'country', 'region', 'state', 'city',
    'attack_type', 'target_type', 'target_subtype',
    'group_name', 'weapon_type', 'weapon_subtype', 
    'property_extent_text'
]

for col in cat_cols:
    gtd[col] = gtd[col].astype('category')
```

```python
bool_cols = [
    'extended_attack', 'multiple_attacks', 'successful_attack', 
    'suicide_attack', 'double_terrorism', 'is_hostage_kidnapping'
]

for col in bool_cols:
    gtd[col] = gtd[col].astype(bool)
```

```python
# Handle missing values for critical columns
gtd = gtd.fillna({
    'num_killed': 0,
    'num_wounded': 0, 
    'num_terrorists_killed': 0,
    'num_terrorists_wounded': 0,
    'num_hostages': 0,
    'group_suspected': 0,
    'property_value': 0,
    'ransom_demanded': 0
})
```

```python
#Convert people counts from float to integer
people_columns = [
    'num_killed', 'num_wounded', 'num_terrorists_killed',
    'num_terrorists_wounded', 'num_hostages'
]
for col in people_columns:
    gtd[col] = gtd[col].astype('int32')  # Use int32 for potential large numbers
```

```python
# Fix the boolean column that's still float
gtd['group_suspected'] = gtd['group_suspected'].astype('bool')
```

```python
# Optimize remaining floats
gtd['property_value'] = gtd['property_value'].astype('float32')
gtd['ransom_demanded'] = gtd['ransom_demanded'].astype('float32')
```

```python
#Let's save some memory if we can

gtd = gtd.apply(
    lambda x: pd.to_numeric(x, downcast='integer') if x.dtype == 'int64' else
              pd.to_numeric(x, downcast='float') if x.dtype == 'float64' else x
)
```

```python
#Checking to make sure it dropped

gtd.info(memory_usage='deep')
```

```python
#Dropping duplicates

gtd.drop_duplicates(inplace=True)
```

```python
#Sorting according to dates

gtd.sort_values(by="date", inplace=True)
gtd.reset_index(drop=True, inplace=True)
```

```python
 #Checking if the NaN counting is valid or filled with replacements like -9, -99, -999 

# Check if any -9, -99, -999 still exist in numeric columns
for col in gtd.select_dtypes(include=['float', 'int']).columns:
    if (gtd[col].isin([-9, -99, -999])).any():
        print(col, gtd[col].isin([-9, -99, -999]).sum())
```

```python
#let's replace them with actual NaNs for proper imputing where it is appropriate

import numpy as np
gtd.replace([-9, -99, -999], np.nan, inplace=True)
```

```python
#I am going to fill the property_value with O meaning no damage or unknown value (no damage done to the dataset here)
#Next num_hostages as we can always cross check with is_hostage_kidnapping
#filling 0 in ransom_demanded is just fine (at least with the analysis we are going to do --disregrding those perspectives where 0 means no ransom_demandeed)

gtd = gtd.fillna({
    'property_value': 0,
    'num_hostages': 0,
    'ransom_demanded': 0
})
```

### 1. Temporal Analysis - The Evolution of Terrorism

```python
#Average per attack types

gtd.groupby('attack_type', observed=True)['num_killed'].mean(numeric_only=True).sort_values(ascending=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Yearly trend of incidents
plt.figure(figsize=(12,6))
sns.countplot(x='year', data=gtd, hue='year', palette='mako', legend=False)
plt.title('Number of Terrorist Attacks per Year')
plt.xticks(rotation=90)
plt.show()
```

> We can see that terrorist attacks have peaked in the year 2014, crossing more than 16000+ attacks that year. 

```python
#Decadal Comparison

# Creating decade column
gtd['decade'] = (gtd['year'] // 10) * 10
decades = sorted(gtd['decade'].unique())
decade_labels = [f"{decade}s" for decade in decades]

plt.figure(figsize=(20, 12))

# Overall trends
plt.subplot(2, 3, 1)
decade_counts = gtd.groupby('decade').size()
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']

bars = plt.bar(decade_labels, decade_counts.values, color=colors, alpha=0.8)
plt.title('Terrorist Incidents by Decade', fontsize=14, fontweight='bold')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=45)
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 100,
             f'{int(height):,}', ha='center', va='bottom', fontweight='bold')
```

```python
# Casuality trends
plt.figure(figsize=(10, 6)) 
decade_casualties = gtd.groupby('decade')[['num_killed', 'num_wounded']].sum()
decade_casualties['total'] = decade_casualties['num_killed'] + decade_casualties['num_wounded']

x = np.arange(len(decade_labels))
width = 0.25

plt.bar(x - width, decade_casualties['num_killed'], width, label='Fatalities', color='#d62728', alpha=0.8)
plt.bar(x, decade_casualties['num_wounded'], width, label='Injured', color='#ff7f0e', alpha=0.8)
plt.bar(x + width, decade_casualties['total'], width, label='Total Casualties', color='#2ca02c', alpha=0.8)

plt.title('Casualties by Decade', fontsize=14, fontweight='bold')
plt.ylabel('Number of People')
plt.xticks(x, decade_labels, rotation=45)
plt.legend()
```

```python
# Success rates
plt.figure(figsize=(10, 6)) 
success_rates = gtd.groupby('decade')['successful_attack'].mean() * 100

plt.plot(decade_labels, success_rates.values, marker='o', linewidth=3, markersize=8, 
         color='#e377c2', markerfacecolor='white', markeredgewidth=2)
plt.title('Success Rates Over Decades', fontsize=10, fontweight='bold')
plt.ylabel('Success Rate (%)')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.ylim(80, 96)

# Value labels on points
for i, rate in enumerate(success_rates.values):
    plt.annotate(f'{rate:.1f}%', (i, rate), textcoords="offset points", 
                 xytext=(0,10), ha='center', fontweight='bold')
```

> Though the 2010s witnessed a surge in terrorist attacks, the success rates were actually higher in the 2000s, averaging around 93%.
>
> Interestingly, the 2020s show a noticeable decline in both the number and success of attacks  possibly showing the impact of advanced surveillance, and other technologies

Between 2014 and 2016, terrorism incidents peaked globally, largely driven by conflicts in the Middle East and Africa. Iraq and Syria saw surges due to the rise of ISIS, while Boko Harams activity intensified in Nigeria. This reflects how organized militant networks during that period led to regional instability and higher attack frequency

```python
# Attack type evolution
plt.figure(figsize=(10, 6)) 
attack_evolution = pd.crosstab(gtd['decade'], gtd['attack_type'], normalize='index') * 100
top_5_attacks = attack_evolution.sum().nlargest(5).index

attack_evolution[top_5_attacks].plot(kind='bar', ax=plt.gca(), width=0.8)
plt.title('Top 5 Attack Methods by Decade', fontsize=14, fontweight='bold')
plt.ylabel('Percentage of Attacks (%)')
plt.xlabel('Decade')
plt.legend(title='Attack Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
```

> Bombings and explosions remain the most frequently used methods, followed by armed assaults.
Interestingly, during the 1970s, assassinations were more common than armed assaults  a reflection of the political motives and targeted violence of that era, though bombings still dominated overall.

```python
# Regional shifts
plt.figure(figsize=(10, 6)) 
region_evolution = pd.crosstab(gtd['decade'], gtd['region'], normalize='index') * 100
top_5_regions = region_evolution.sum().nlargest(5).index

region_evolution[top_5_regions].plot(kind='area', ax=plt.gca(), alpha=0.7, cmap='viridis')
plt.title('Regional Distribution Over Decades', fontsize=14, fontweight='bold')
plt.ylabel('Percentage of Attacks (%)')
plt.xlabel('Decade')
plt.legend(title='Region', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
```

> Sub-Saharan Africa exhibits a significantly greater concentration of attacks than any other region, probably due to its economic vulnerabilities, political instability, and even ethnic tensions.

```python
# Lethality trends
plt.figure(figsize=(10, 6)) 
lethality = gtd.groupby('decade')['num_killed'].mean()
casualty_rate = (gtd['num_killed'] + gtd['num_wounded']).groupby(gtd['decade']).mean()

x = np.arange(len(decade_labels))
plt.bar(x - 0.2, lethality.values, 0.4, label='Avg Deaths/Attack', color='#c44e52', alpha=0.8)
plt.bar(x + 0.2, casualty_rate.values, 0.4, label='Avg Total Casualties/Attack', color='#55a868', alpha=0.8)

plt.title('Attack Lethality Over Decades', fontsize=14, fontweight='bold')
plt.ylabel('Average per Attack')
plt.ylim(0, max(casualty_rate.values) * 1.2)  

plt.xticks(x, decade_labels, rotation=45)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)

plt.tight_layout()
plt.show()
```

> This graph is a proof to our previous conclusion -- though 2000s saw more terrorist attacks, the actual lethality was seen in 1980s with more fatalities per attack.

```python
# Calculating key metrics
current_decade = decades[-1]
previous_decade = decades[-2]

current_incidents = decade_counts[current_decade]
previous_incidents = decade_counts[previous_decade]
growth_pct = ((current_incidents - previous_incidents) / previous_incidents) * 100

most_dangerous_decade = decade_counts.idxmax()
most_dangerous_region = gtd.groupby('decade')['region'].apply(lambda x: x.mode()[0])

# Top attack method for each decade
print("DOMINANT ATTACK METHODS BY DECADE:")
for decade in decades:
    top_attack = gtd[gtd['decade'] == decade]['attack_type'].value_counts().index[0]
    print(f"   {decade}s: {top_attack}")
```

### Monthly Analysis

```python
# Monthly analysis
monthly_attacks = gtd['month'].value_counts().sort_index()
monthly_attacks = monthly_attacks[(monthly_attacks.index >= 1) & (monthly_attacks.index <= 12)]

monthly_casualties = gtd.groupby('month')[['num_killed', 'num_wounded']].sum()
monthly_casualties = monthly_casualties[(monthly_casualties.index >= 1) & (monthly_casualties.index <= 12)]

# Month names for better readability
month_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
               'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

plt.figure(figsize=(15, 10))

# Monthly attack frequency
plt.subplot(2, 2, 1)
bars = plt.bar(month_names, monthly_attacks.values, color='skyblue', alpha=0.8)
plt.title('Terrorist Attacks by Month', fontsize=14, fontweight='bold')
plt.ylabel('Number of Attacks')
plt.grid(True, alpha=0.3)

# Highlight peak month
peak_month = monthly_attacks.idxmax()
plt.axvline(x=peak_month-1, color='red', linestyle='--', alpha=0.7, 
            label=f'Peak: {month_names[peak_month-1]}')
plt.legend()
plt.tight_layout()
plt.show()
```

> We notice more attacks in the month of May compared to other months  possibly due to changes in political cycles such as elections or shifts in leadership that often occur mid-year. Other reasons could include religious reasons, ideal weather especially in areas like South Asia, Africa etc.

```python
# Monthly casualties
plt.figure(figsize=(10,6))
plt.plot(month_names, monthly_casualties['num_killed'], marker='o', linewidth=2, 
         label='Fatalities', color='red')
plt.plot(month_names, monthly_casualties['num_wounded'], marker='s', linewidth=2, 
         label='Injured', color='orange')
plt.title('Casualties by Month', fontsize=14, fontweight='bold')
plt.ylabel('Number of People')
plt.legend()
plt.grid(True, alpha=0.3)
```

> Interestingly, though May have more attacks, September seem to see far more injuries than other months. It could be due to far deadlier attacks that period.

```python
# Success rates by month
plt.figure(figsize=(10,6))
monthly_success = gtd.groupby('month')['successful_attack'].mean() * 100

# Filter only months 1-12 and sort properly
monthly_success = monthly_success[(monthly_success.index >= 1) & (monthly_success.index <= 12)]
monthly_success = monthly_success.sort_index()

# Month names for readability
month_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
               'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

plt.plot(month_names, monthly_success.values, marker='o', color='green', linewidth=2)
plt.title('Success Rates by Month', fontsize=14, fontweight='bold')
plt.ylabel('Success Rate (%)')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

> Success rates of attacks show minimal seasonal variation, suggesting month is not a strong influencing factor.

```python
#Seasonal Patterns (quarters)
gtd_valid = gtd[gtd['month'].between(1, 12)].copy()
plt.figure(figsize=(6,6))
gtd_valid['quarter'] = (gtd_valid['month'] - 1) // 3 + 1 

quarterly_attacks = gtd_valid.groupby('quarter').size()
quarters = ['Q1 (JanMar)', 'Q2 (AprJun)', 'Q3 (JulSep)', 'Q4 (OctDec)']

plt.pie(
    quarterly_attacks.values,
    labels=quarters,
    autopct='%1.1f%%',
    colors=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']
)
plt.title('Attack Distribution by Quarter', fontsize=14, fontweight='bold')
plt.show()
```

```python
# Most active terrorist groups (other than Unknown)
plt.figure(figsize=(10, 6))
group_activity = gtd['group_name'].value_counts().head(10)

# Filter out "Unknown" if it's in top groups
if 'Unknown' in group_activity.index:
    group_activity = group_activity[group_activity.index != 'Unknown'].head(10)

bars = plt.barh(range(len(group_activity)), group_activity.values, 
                color=plt.cm.Reds(np.linspace(0.4, 1, len(group_activity))))
plt.yticks(range(len(group_activity)), group_activity.index)
plt.title('Top 10 Most Active Terrorist Groups', fontsize=16, fontweight='bold')
plt.xlabel('Number of Attacks')

# Add value labels
for i, bar in enumerate(bars):
    width = bar.get_width()
    plt.text(width + 50, bar.get_y() + bar.get_height()/2, 
             f'{int(width):,}', ha='left', va='center', fontweight='bold')
```

```python
# Group Lifespan and their activity levels
plt.figure(figsize=(10,6))

# activity periods of top groups
top_groups = group_activity.index[:5]
group_years = {}

for group in top_groups:
    group_data = gtd[gtd['group_name'] == group]
    active_years = group_data['year'].unique()
    group_years[group] = (active_years.min(), active_years.max(), len(active_years))

# timeline visualization
fig, ax = plt.subplots(figsize=(12, 6))
years_range = range(gtd['year'].min(), gtd['year'].max() + 1)

for i, (group, (start, end, span)) in enumerate(group_years.items()):
    ax.plot([start, end], [i, i], linewidth=8, label=f'{group} ({span} years)')
    ax.text(start, i, f' {group}', va='center', fontsize=9)

ax.set_yticks(range(len(group_years)))
ax.set_yticklabels([group.split()[0] for group in group_years.keys()])  # Short names
ax.set_xlabel('Year')
ax.set_title('Activity Timeline of Top 5 Terrorist Groups', fontsize=16, fontweight='bold')
ax.grid(True, alpha=0.3)
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
```

### 5. Impact Assessment - Human & Economic Cost

- Casualty trends over time
-  High-casualty event analysis
- Property damage patterns
- Regional impact comparisons

```python
# Casuality over time
plt.figure(figsize=(10,6))

# yearly casualties
yearly_casualties = gtd.groupby('year').agg({
    'num_killed': 'sum',
    'num_wounded': 'sum'
})
yearly_casualties['total_casualties'] = yearly_casualties['num_killed'] + yearly_casualties['num_wounded']

#area chart
plt.stackplot(yearly_casualties.index, 
              yearly_casualties['num_killed'], 
              yearly_casualties['num_wounded'],
              labels=['Fatalities', 'Injured'],
              colors=['#c44e52', '#4c72b0'], alpha=0.8)

plt.title('Casualty Trends Over Time (1970-2020)', fontsize=16, fontweight='bold')
plt.ylabel('Number of Casualties')
plt.xlabel('Year')
plt.legend(loc='upper left')
plt.grid(True, alpha=0.3)
```

```python
# High casuality events
plt.figure(figsize=(10,6))

#top 1% by casualties
casualty_threshold = gtd['total_casualties'] = gtd['num_killed'] + gtd['num_wounded']
high_casualty_threshold = casualty_threshold.quantile(0.99)

high_casualty_events = gtd[casualty_threshold >= high_casualty_threshold]

#high-casualty events by attack type
hc_by_type = high_casualty_events['attack_type'].value_counts().head(8)

bars = plt.barh(range(len(hc_by_type)), hc_by_type.values,
                color=plt.cm.Reds(np.linspace(0.4, 1, len(hc_by_type))))
plt.yticks(range(len(hc_by_type)), hc_by_type.index)
plt.title(f'High-Casualty Events by Attack Type\n(Top 1%: {high_casualty_threshold:.0f}+ casualties)', 
          fontsize=16, fontweight='bold')
plt.xlabel('Number of High-Casualty Events')

# Add value labels
for i, bar in enumerate(bars):
    width = bar.get_width()
    plt.text(width + 0.5, bar.get_y() + bar.get_height()/2,
             f'{int(width)}', ha='left', va='center', fontweight='bold')
```

```python
#Property damage
plt.figure(figsize=(10,6))

# Analyze property damage
property_analysis = gtd.groupby('attack_type').agg({
    'property_damage': 'mean',
    'property_value': lambda x: x[x > 0].mean()
}).rename(columns={'property_damage': 'damage_frequency', 'property_value': 'avg_damage_amount'})

property_analysis = property_analysis.sort_values('avg_damage_amount', ascending=False).head(8)

# dual-axis plot
ax1 = plt.gca()
ax2 = ax1.twinx()

x = np.arange(len(property_analysis))
width = 0.35

# Damage frequency (left axis)
bars1 = ax1.bar(x - width/2, property_analysis['damage_frequency'] * 100, width,
                color='skyblue', alpha=0.8, label='Damage Frequency')
# Average damage amount (right axis)
bars2 = ax2.bar(x + width/2, property_analysis['avg_damage_amount'] / 1000000, width,
                color='coral', alpha=0.8, label='Avg Damage (Millions $)')

ax1.set_title('Property Damage Patterns by Attack Type', fontsize=16, fontweight='bold')
ax1.set_ylabel('Damage Frequency (%)', color='skyblue')
ax2.set_ylabel('Average Damage (Millions $)', color='coral')
ax1.set_xticks(x)
ax1.set_xticklabels([name[:15] + '...' if len(name) > 15 else name for name in property_analysis.index], 
                    rotation=45, ha='right')

# Add legends
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')
```

```python
# Regional Impacts
plt.figure(figsize=(10,6))

# regional impact
regional_impact = gtd.groupby('region').agg({
    'num_killed': 'sum',
    'num_wounded': 'sum',
    'property_value': 'sum',
    'eventid': 'count'
}).rename(columns={'eventid': 'incident_count'})

regional_impact['total_casualties'] = regional_impact['num_killed'] + regional_impact['num_wounded']
regional_impact['casualties_per_attack'] = regional_impact['total_casualties'] / regional_impact['incident_count']
regional_impact['damage_per_attack'] = regional_impact['property_value'] / regional_impact['incident_count']

# Normalizing for radar chart-like comparison (0-1 scale)
for col in ['casualties_per_attack', 'damage_per_attack', 'incident_count']:
    regional_impact[f'{col}_norm'] = regional_impact[col] / regional_impact[col].max()

#top 6 regions
top_regions = regional_impact.nlargest(6, 'total_casualties')
metrics = ['casualties_per_attack_norm', 'damage_per_attack_norm', 'incident_count_norm']
metric_labels = ['Casualty Intensity', 'Damage Intensity', 'Attack Frequency']

x = np.arange(len(metrics))
width = 0.15

for i, (region, data) in enumerate(top_regions.iterrows()):
    plt.bar(x + i*width, [data[metric] for metric in metrics], width, label=region)

plt.title('Regional Impact Comparison (Normalized)', fontsize=16, fontweight='bold')
plt.ylabel('Normalized Impact Score (0-1)')
plt.xticks(x + width*2.5, metric_labels)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, alpha=0.3, axis='y')

plt.tight_layout(pad=3.0)
plt.show()
```

Terrorism has evolved drastically over the past five decades  shifting in geography, strategy, and scale. The 2000s witnessed the highest surge in attacks, largely fueled by global conflicts and ideological extremism, while the 1980s stand out as the most lethal period per incident.

Bombings remain the dominant form of attack, yet the use of firearms and explosives has diversified.

Although recent decades show a decline in total incidents, the persistence of organized terror networks highlights that modern terrorism is far from eradicated  it has merely adapted to new forms, exploiting digital communication, propaganda, and technology.
The overall downward trend in the 2020s, however, may reflect improved intelligence, counter-terrorism measures, and global cooperation  marking a cautious optimism for the future.
